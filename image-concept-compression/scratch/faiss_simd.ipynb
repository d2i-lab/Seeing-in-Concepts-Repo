```python
import pandas as pd

df = pd.read_csv('benchmark_results.csv')
df
```

```python
import matplotlib.pyplot as plt

# Plot relationship between OMP_NUM_THREADS and queries per second
plt.figure(figsize=(10, 6))


# Plot relationship between avg_points_considered and queries per second
# separate lines for each OMP_NUM_THREADS
plt.figure(figsize=(10, 6))
for threads in df['OMP_NUM_THREADS'].unique():
    df_threads = df[df['OMP_NUM_THREADS'] == threads]
    print(df_threads)
    plt.plot(df_threads['total_points'], df_threads['runtime'], label=f'Threads: {threads}', marker='o')
plt.xlabel('Total Points Considered')
plt.ylabel('Runtime (s)')
plt.title('Total Points Considered vs Runtime')
plt.legend()


```
